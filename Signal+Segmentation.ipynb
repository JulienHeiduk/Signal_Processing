```python
import pandas as pd
import matplotlib.pyplot as plt
import ruptures as rpt

data = pd.read_csv('C:/Users/jheiduk/Downloads/DJENA_Elan.txt', sep='\t' )
y  = data.AX[:10000]

del data
```

```python
# detection
algo = rpt.Pelt(model="rbf").fit(y)
result = algo.predict(pen = 20)

# display
#plt.plot(y)
#plt.show()
```

```python
# display
rpt.display(y, result, result)
plt.show()
```

```python
result_bis = []

for i in range(len(result)):
    if i > 0:
        result_bis.append(result[i] - result[i-1])
        
#result_bis
```

```python
import numpy as np
arr = np.array(result_bis)
indx = np.where(arr > 250)[0]
```

```python
result_out = [result[i] for i in indx]
#result_out
```

```python
# display
rpt.display(y, result_out, result_out)
plt.show()
```
